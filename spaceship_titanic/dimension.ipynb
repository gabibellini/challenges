# Feature Engineering

```python
# função para dizer se a feature foi ou não processada

def status(feature):
    print('Processando', feature, ': ok')
```

```python
# unindo train e test para previnir vazamento de informações e adicionar novas variáveis

def combined_data():

    # lendo dados de treino
    train = pd.read_csv('train.csv')

    # lendo dados de teste
    test = pd.read_csv('test.csv')

    # extração e remoção da variável alvo do conjunto de treino
    targets = train.Transported
    train.drop(['Transported'], axis = 1, inplace = True)

    # merge os dois conjuntos para feature engineering
    combined = pd.concat([train, test])
    combined.reset_index(inplace = True)
    combined.drop(['index'], axis = 1, inplace = True)

    return combined

combined = combined_data()
```

```python
combined.shape
```

```python
combined.head(10)
```

```python
# extraindo grupos e tamanho do grupo

def grupo_tamanho(combined):

    # extração do grupo de cada passanger id
    combined['Group'] = combined['PassengerId'].map(lambda pid: pid.split('_')[0].strip())
    combined['Size'] = combined['PassengerId'].map(lambda pid: pid.split('_')[1].strip())

    # convertendo para int
    combined['Group'] = combined.Group.astype(int)
    combined['Size'] = combined.Size.astype(int)
    combined['MaxSize'] = combined.groupby('Group')['Size'].transform('max')
    combined.drop('Size', axis = 1, inplace = True)
    combined.rename(columns = {'MaxSize': 'Size'}, inplace = True)

    return combined
```

```python
combined = grupo_tamanho(combined)
```

```python
combined.head()
```

```python
# extraindo deck

def cabine(combined):

    combined['Cabin'].fillna('U', inplace = True)

    # extração do grupo de cada cabine
    combined['Deck'] = combined['Cabin'].map(lambda cabin: cabin.split('/')[0].strip() if cabin != 'U' else 'U')
    combined['Number'] = combined['Cabin'].map(lambda cabin: cabin.split('/')[1].strip() if cabin != 'U' else 'U')
    combined['Side'] = combined['Cabin'].map(lambda cabin: cabin.split('/')[2].strip() if cabin != 'U' else 'U')
            

    # convertendo number para int
    combined['Number'] = combined['Number'].replace('U', '-1').astype(int) + 1

    return combined
```

```python
combined = cabine(combined)
combined.head(10)
```

```python
def preenche_gastos(combined):
    combined.RoomService.fillna(0, inplace = True)
    combined.FoodCourt.fillna(0, inplace = True)
    combined.ShoppingMall.fillna(0, inplace = True)
    combined.Spa.fillna(0, inplace = True)
    combined.VRDeck.fillna(0, inplace = True)

    return combined
```

```python
# gastos totais

def gastos_totais(combined):
    combined['Total'] = combined['RoomService'] + combined['FoodCourt'] + combined['ShoppingMall'] + combined['Spa'] + combined['VRDeck']

    return combined
```

```python
combined = gastos_totais(preenche_gastos(combined))
combined.head()
```

```python
def processa_homeplanet(combined):
    
    # tem apenas dois valores faltantes, preencher com o valor mais frequente no dataset de treino
    combined.HomePlanet.fillna('Earth', inplace = True)

    # dummy enconding
    homeplanet_dummies = pd.get_dummies(combined['HomePlanet'], prefix = 'HomePlanet', dtype = int)
    combined = pd.concat([combined, homeplanet_dummies], axis = 1)
    combined.drop('HomePlanet', axis = 1, inplace = True)

    status('homeplanet')
    return combined
```

```python
combined = processa_homeplanet(combined)
combined.head()
```

```python
combined.Destination.value_counts()
```

```python
def processa_destination(combined):
    
    # tem apenas dois valores faltantes, preencher com o valor mais frequente no dataset de treino
    combined.Destination.fillna('TRAPPIST-1e', inplace = True)

    # dummy enconding
    destination_dummies = pd.get_dummies(combined['Destination'], prefix = 'Destination', dtype = int)
    combined = pd.concat([combined, destination_dummies], axis = 1)
    combined.drop('Destination', axis = 1, inplace = True)

    status('destination')
    return combined
```

```python
train_group = combined.iloc[:8693].groupby(['Size', 'Deck', 'VIP'])
train_group_median = train_group.Age.median()
train_group_median = train_group_median.reset_index()[['Size', 'Deck', 'VIP', 'Age']]

train_group_median
```

```python
# criando a função que preenche os valores faltantes de acordo com o critério acima

def preenche_idade(linha):
    condicao = (
        (train_group_median['Size'] == linha['Size']) &
        (train_group_median['Deck'] == linha['Deck']) &
        (train_group_median['VIP'] == linha['VIP'])
    )
    return train_group_median[condicao]['Age'].values[0]

def processa_idade(combined):

    # função que preenche os valores faltantes da coluna Age
    combined['Age'] = combined.apply(lambda linha: preenche_idade(linha) if np.isnan(
        linha['Age']) else linha['Age'], axis = 1)
    
    status('age')
    return combined
```

```python
combined = processa_idade(combined)
combined.info()
```

```python
combined.isnull().sum()
```

```python
def processa_bools(combined):
    combined['CryoSleep'] = combined['CryoSleep'].replace({'True': 1, 'False': 0}).astype(int)
    combined['VIP'] = combined['VIP'].replace({'True': 1, 'False': 0}).astype(int)

    return combined   
```

```python
def processa_dummies(combined):
    class_dummies = pd.get_dummies(combined[['Deck']], prefix = 'Deck', dtype = int)
    side_dummies = pd.get_dummies(combined['Side'], prefix = 'Side', dtype = int)
    combined = pd.concat([combined, side_dummies, class_dummies], axis = 1)
    combined.drop(['Deck', 'Side', 'Name', 'PassengerId', 'Cabin'], axis = 1, inplace = True)

    status('classe')
    return combined
```

```python
def processa_grupo(combined):
    
    # novas features baseadas no tamanho da família
    combined['Singleton'] = combined['Size'].map(lambda s: 1 if s == 1 else 0)
    combined['SmallGroup'] = combined['Size'].map(lambda s: 1 if 2 <= s <= 4 else 0)
    combined['LargeGroup'] = combined['Size'].map(lambda s: 1 if 5 <= s else 0)
    
    status('grupo')
    return combined
```

```python
combined = processa_bools(combined)
combined = processa_dummies(combined)
combined = processa_grupo(combined)
combined.head()
```

# modelo

```python
def recover_train_test(combined):

    y = pd.read_csv('train.csv', usecols = ['Transported'])['Transported'].values
    train = combined.iloc[:8693]
    test = combined.iloc[8693:]

    return y, train, test
```

```python
y, train, test = recover_train_test(combined)
```

```python
from imblearn.over_sampling import SMOTE

SEED = 304
np.random.seed(SEED)

smote = SMOTE(random_state=SEED)

train, y = smote.fit_resample(train, y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier
```

```python
SEED = 304
np.random.seed(SEED)

X_train, X_test, y_train, y_test = train_test_split(train, y, test_size=0.3, stratify = y)
print("Training with %d elements and testing with %d elements." % (len(X_train), len(X_test)))
```

```python
def calcula_metricas(test_y, predictions):
    
    print('Acurácia:', round(metrics.accuracy_score(test_y, predictions)*100, 2))
    print('Precisão:', round(metrics.precision_score(test_y, predictions)*100, 2))
    print('Recall:', round(metrics.recall_score(test_y, predictions)*100, 2))
    print('F1:', round(metrics.f1_score(test_y, predictions)*100, 2))
```

```python
def matriz_confusao(test_y, predictions):

    cm = confusion_matrix(test_y, predictions)
    disp = ConfusionMatrixDisplay(confusion_matrix = cm)
    disp.plot()
```

```python
# dummy
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier()
dummy.fit(X_train, y_train)

acuracia_dummy = dummy.score(X_test, y_test) * 100

print("Acurácia dummy foi: %.2f%%" % acuracia_dummy)
```

```python
# random forest

SEED = 0
np.random.seed(SEED)
rfc = RandomForestClassifier(max_depth=5, n_estimators = 50, max_features= 'sqrt')
rfc.fit(X_train, y_train)
preds = rfc.predict(X_test)

calcula_metricas(y_test, preds)
matriz_confusao(y_test, preds)
```

```python
features = pd.DataFrame()
features['feature'] = train.columns
features['importance'] = rfc.feature_importances_
features.sort_values(by=['importance'], ascending=True, inplace=True)
features.set_index('feature', inplace=True)

features.plot(kind='barh', figsize=(25, 25))
```

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold

SEED = 0
np.random.seed(SEED)

cv = KFold(n_splits = 10, shuffle = True)
model = RandomForestClassifier(max_depth=5)
results = cross_validate(model, train, y, cv = cv, return_train_score= False)

mean = results['test_score'].mean()
std_dev = results['test_score'].std()
print("Accuracy mean: %.2f" % (mean * 100))
print("Accuracy with cross validation = [%.2f, %.2f]" % ((mean - 2 * std_dev) * 100, (mean + 2 * std_dev) * 100))
```

```python
# lista de parâmetros que serão testados

n_estimators = [100, 150, 200]
max_depth = [2, 5, 10]
min_samples_split = [2, 4, 6]
min_samples_leaf = [1, 2, 4]
```

```python
param_grid = {
    'n_estimators' : n_estimators,
    'max_depth' : max_depth,
    'min_samples_split' : min_samples_split,
    'min_samples_leaf' : min_samples_leaf
}
```

```python
from sklearn.model_selection import GridSearchCV

SEED = 0
np.random.seed(SEED)

grid_cv = GridSearchCV(model, param_grid,
                             cv=cv, scoring="accuracy", n_jobs = -1)
```

```python
# rodando o modelo com o grid search
busca = grid_cv.fit(train, y)
```

```python
# cria um dataframe com os resultados do grid search
pd.DataFrame(busca.cv_results_)
```

```python
# encontra os melhores parâmetros
busca.best_params_
```

```python
# rodando de novo o modelo com os parâmetros encontrados

rf = RandomForestClassifier(**busca.best_params_)

modelo = rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
```

```python
calcula_metricas(y_test, y_pred)
matriz_confusao(y_test, y_pred)
```

```python
y_pred_proba = modelo.predict_proba(X_test)[::, 1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)

plt.figure(figsize=(12, 8))
plt.plot(fpr, tpr, label = 'RFC, auc ='+str(auc))
plt.plot([0, 1], [0, 1], color = 'r', lw = 2, linestyle = '--')
plt.legend(loc = 4)
```
